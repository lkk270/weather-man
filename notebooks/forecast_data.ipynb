```python
# First cell - Setup
import sys
from pathlib import Path
import nest_asyncio
import asyncio

# Enable async support
nest_asyncio.apply()

# Add the project root directory to Python path
project_root = str(Path().absolute().parent)
if project_root not in sys.path:
    sys.path.append(project_root)
```

```python
# Second cell - Fetch data
from providers.nws.forecast.api_client import fetch_forecast_data

# Test fetching weather data
location_id = "NYC"  # Using location ID from settings
data = fetch_forecast_data(location_id)

# Print some basic info about the response
print("Raw API Response:")
print("Properties:", data.get('properties'))
if 'properties' in data:
    print("Periods:", len(data['properties'].get('periods', [])))
else:
    print("No properties found in data")
print("---")
```

```python
# Add this cell after fetching data
print("Raw API Response:")
print("Properties:", data.get('properties'))
if 'properties' in data:
    print("Periods:", len(data['properties'].get('periods', [])))
else:
    print("No properties found in data")
print("---")
```

```python
# Third cell - Clean and load data
from providers.nws.forecast import clean_forecast_data, load_forecast_data

try:
    print("About to start cleaning...")
    cleaned_data = clean_forecast_data(data, location_id="NYC")
    print(f"Cleaning complete. Got {len(cleaned_data)} records")
    
    print("Starting database load...")
    await load_forecast_data(cleaned_data)
    print(f"Successfully loaded {len(cleaned_data)} weather records")
except Exception as e:
    print(f"Error occurred: {str(e)}")
    import traceback
    traceback.print_exc()
```
